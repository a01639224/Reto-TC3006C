## Crear función predictora

```
# Definir función
def make_prediction(classifier, back_x, back_y, back_z, thigh_x, thigh_y, thigh_z):
    if(classifier == 'K-Nearest Neighbors'):
        with open('/content/drive/MyDrive/Ciencia de datos e IA/k_nearest_neighbors_model.pkl', 'rb') as f:
            clf  = pickle.load(f)
    if(classifier == 'Random Forest'):
        with open('/content/drive/MyDrive/Ciencia de datos e IA/random_forest_model.pkl', 'rb') as f:
            clf  = pickle.load(f)
    preds = clf.predict([[back_x, back_y, back_z, thigh_x, thigh_y, thigh_z]])
    if preds == 1:
        return 1, 'walking'
    elif preds.tolist() == 3:
        return 3, 'shuffling'
    elif preds == 4:
        return 4, 'stairs (ascending)'
    elif preds == 5:
        return 5, 'stairs (descending)'
    elif preds == 6:
        return 6, 'standing'
    elif preds == 7:
        return 7, 'sitting'
    elif preds == 8:
        return 8, 'lying'
```

## Crear interface

```
# Definir entradas
classifier = gr.Dropdown(['K-Nearest Neighbors', 'Random Forest'], label='classifier')
back_x = gr.Number(label='back_x')
back_y = gr.Number(label='back_y')
back_z = gr.Number(label='back_z')
thigh_x = gr.Number(label='thigh_x')
thigh_y = gr.Number(label='thigh_y')
thigh_z = gr.Number(label='thigh_z')

# Definir salidas
act_class = gr.Number(label = 'class')
activity = gr.Text(label = 'activity')

# Crear interface
app = gr.Interface(fn=make_prediction, inputs=[classifier, back_x, back_y, back_z, thigh_x, thigh_y, thigh_z], outputs=[act_class, activity]).launch(share=True)
```
